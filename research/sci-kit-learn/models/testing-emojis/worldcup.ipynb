# Loading the Twitter Data

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', None)
data = pd.read_csv("./worldcup.csv")
data.head()
```

```python
from emoji import UNICODE_EMOJI

def extractEmojis(text):
    for emoji in UNICODE_EMOJI['en']:
        text = text.replace(emoji ,"")

    return text

data['comment_text'] = data['comment_text'].apply(extractEmojis)
data.head()
```

# Text Preprocessing

```python
# Convert comment to lowercase
def to_lowercase(text):
    return text.lower()

data['comment_text'] = data['comment_text'].apply(to_lowercase)
data.head()
```

```python
import re
# Remove HTML tags from the comments
def remove_html(text):
    return re.sub(r"<.*>", "", text, flags=re.MULTILINE)
    
data['comment_text'] = data['comment_text'].apply(remove_html)
data.head()
```

```python
# Remove links from the comments
def remove_links(text):
    text= re.sub(r"http\S+"," ",text, flags=re.MULTILINE)
    return re.sub(r"www\S+"," ",text, flags=re.MULTILINE)

data['comment_text'] = data['comment_text'].apply(remove_links)
data.head()
```

```python
import string
# Remove punctuation marks 
def remove_punctuation(text):
    for i in string.punctuation:
        text = text.replace(i, "")
    return text

data['comment_text'] = data['comment_text'].apply(remove_punctuation)
data.head()
```

```python
# Remove special characters such as: \n \r \t
def remove_special(text):
    return re.sub(r"[\n\t\\\/\r]"," ",text, flags=re.MULTILINE)

data['comment_text'] = data['comment_text'].apply(remove_special)
data.head()
```

```python
# Remove stopwords using nltk's stopwords package
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

data['comment_text'] = data['comment_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
data.head()
```

```python
# As you can see above, there are numbers and/or dates
# I will remove those as they are not helpful

def remove_numbers(text):
    return re.sub(r'\d'," ",text, flags=re.MULTILINE)

data['comment_text'] = data['comment_text'].apply(remove_numbers)
data.head()
```

# Loading the Saved Models

```python
import joblib


labels = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
models = {}

for label in labels:
    models[label] = joblib.load(open(f'F:/Thesis/models/logistic/{label}.pkl', 'rb'))
```

```python
data = data.sort_values(by=labels, ascending=False)
data.head()
```

```python
data.head()
```

```python
# Select only the rows which have a 1 in one of it's toxic columns
final_data = data[data.isin([1]).any(axis=1)]
final_data.head(25)
```

```python
from emoji import UNICODE_EMOJI

def extractEmojis(text):
    emojis = []

    if type(text) != str:
        return []
        
    for emoji in text.split(","):
        emojis.append(emoji)

    return emojis
```

```python
# Convert emojiset from string to list of strings
final_data["new_emojisets"] = final_data["emojiset"].apply(extractEmojis)
final_data.head(25)
```

# Frequent Itemset Mining

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
tr = TransactionEncoder()
tr_arr = tr.fit(final_data['new_emojisets']).transform(final_data['new_emojisets'])
df = pd.DataFrame(tr_arr, columns=tr.columns_)
df.head()
```

Below is the first 10 frequent itemsets that were found based on the emojis.

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)
frequent_itemsets.head(50)
```

# Association Rule Mining

Below is the first 10 association rules that were found from the frequent itemsets.

```python
association_rules = association_rules(frequent_itemsets)
association_rules.head(50)
```
